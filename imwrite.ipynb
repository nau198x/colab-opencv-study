```python
import cv2
#サポートパッチのインポート
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/My Drive/Colab Notebooks/img/lena.jpg', cv2.IMREAD_UNCHANGED)
```

```python
_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # とりあえずグレースケール化
```

```python
cv2_imshow(_img)
```

```python
cv2.imwrite('/content/drive/My Drive/Colab Notebooks/img/lena_gray.jpg', _img)
```
